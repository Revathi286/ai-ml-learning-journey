Reflection

When starting with Grid Search, using a wider range of values like [0.1, 1, 10, 100] is helpful because it gives a bigger picture of how the model behaves across different settings.
It helps us quickly figure out which range of values works best. Once we find that, we can then try smaller steps (like 1.1, 1.2, 1.3) to fine-tune the model and get even better performance.
If we start directly with small increments, we might focus on the wrong area and completely miss the best values, which can lead to a less effective model.
