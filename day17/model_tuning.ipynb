DAY 17 : HYPERPARAMETER TUNING

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
X, y = data.data, data.target
```

```python
param_grid = {
'n_estimators': [50, 100, 200],
'max_depth': [None, 10, 20],
'min_samples_split': [2, 5, 10]
}
```

```python


rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
cv=3, scoring='accuracy', verbose=1)
```

```python
grid_search.fit(X, y)
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Cross-Validation Score:{grid_search.best_score_:.4f}")
```

```python
import joblib
best_model = grid_search.best_estimator_
joblib.dump(best_model, "best_model.pkl")

print("Model saved as best_model.pkl")
```

```python
total_combinations = (
    len(param_grid['n_estimators']) *
    len(param_grid['max_depth']) *
    len(param_grid['min_samples_split'])
)
total_fits = total_combinations * 3  

print("Total parameter combinations:", total_combinations)
print("Total model fits (with CV):", total_fits)
```

Q1: How many models will GridSearch train?

If:5 parameters
10 values each
Then:10^5=100000 combinations

With 3-fold CV:
100000×3=300000 model trainings

RandomizedSearchCV is a better choice for large datasets because it doesn’t try every single combination of hyperparameters like GridSearchCV, which can take a lot of time and computing power. Instead, it randomly picks a few combinations to test, making the process much faster and more efficient. Even though it doesn’t check everything, it usually still finds very good (near-optimal) results, making it a practical option when working with big data.
