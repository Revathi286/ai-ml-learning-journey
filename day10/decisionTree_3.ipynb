Day 10: Non-Linear Models & Overfitting.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

X = 6 * np.random.rand(100, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(100, 1)

X_new = np.linspace(-3, 3, 100).reshape(100, 1)

tree_reg_2 = DecisionTreeRegressor(max_depth=2)
tree_reg_2.fit(X, y)
y_pred_2 = tree_reg_2.predict(X_new)
plt.plot(X_new, y_pred_2, color='green', label='Depth=2')


tree_reg_5 = DecisionTreeRegressor(max_depth=5)
tree_reg_5.fit(X, y)
y_pred_5 = tree_reg_5.predict(X_new)
plt.plot(X_new, y_pred_5, color='orange', label='Depth=5')

tree_reg_20 = DecisionTreeRegressor(max_depth=20)
tree_reg_20.fit(X, y)
y_pred_20 = tree_reg_20.predict(X_new)
plt.plot(X_new, y_pred_20, color='red', label='Depth=20')

plt.scatter(X, y, color='blue', label='Data')
plt.legend()
plt.show()
```

Why is a 'jittery' model that hits every training point actually worse for making future predictions than a smooth 'curvy' model?

A jittery model isn’t actually better just because it matches every data point. It ends up learning the noise instead of the real pattern, which is called overfitting. Because of this, it may look perfect on training data but performs poorly on new data.
In contrast, a smooth curvy model focuses on the overall trend rather than every small fluctuation. This makes it more reliable and better at predicting unseen data.
So, a model that generalizes well is more useful than one that simply memorizes the training data.
