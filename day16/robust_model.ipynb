Day 16: Robust Model Evaluation (Shuffle Test)

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_digits
import numpy as np


digits = load_digits()
X, y = digits.data, digits.target
```

```python
model = RandomForestClassifier(n_estimators=50, random_state=42)
```

```python
scores = cross_val_score(model, X, y, cv=5)

print("5-Fold Cross Validation Scores:")
for i, score in enumerate(scores):
    print(f"Fold {i+1}: {score:.4f}")

mean_score = np.mean(scores)
std_score = np.std(scores)

print(f"\nMean Accuracy: {mean_score:.4f}")
print(f"Standard Deviation (Stability): {std_score:.4f}")
```

```python
model.fit(X, y)
train_score = model.score(X, y)

print("\nOverfitting Analysis:")
print(f"Training Accuracy: {train_score:.4f}")
print(f"Validation Accuracy: {mean_score:.4f}")

if train_score > mean_score + 0.1:
    print("Model is Overfitting (High variance)")
elif train_score < mean_score:
    print("Model is Underfitting (High bias)")
else:
    print("Model is Generalizing Well")
```

```python
# Without Shuffling
kf_no_shuffle = KFold(n_splits=10, shuffle=False)
scores_no_shuffle = cross_val_score(model, X, y, cv=kf_no_shuffle)


# With Shuffling
kf_shuffle = KFold(n_splits=10, shuffle=True, random_state=42)
scores_shuffle = cross_val_score(model, X, y, cv=kf_shuffle)

print("\nShuffle Test Results:")

print("\nWithout Shuffling:")
for i, score in enumerate(scores_no_shuffle):
    print(f"Fold {i+1}: {score:.4f}")
print(f"Standard Deviation: {np.std(scores_no_shuffle):.4f}")

print("\nWith Shuffling:")
for i, score in enumerate(scores_shuffle):
    print(f"Fold {i+1}: {score:.4f}")
print(f"Standard Deviation: {np.std(scores_shuffle):.4f}")
```

```python
print("\nStability Comparison:")

if np.std(scores_shuffle) < np.std(scores_no_shuffle):
    print("Shuffling improves model stability")
else:
    print("Shuffling did not improve stability significantly")
```

Observation:

The standard deviation of the model’s accuracy without shuffling is higher compared to when shuffling is applied.This shows that without shuffling, the model performance varies more across folds, making it less stable.When shuffling is enabled, the data is randomly distributed, resulting in more consistent performance and lower standard deviation.
Therefore, shuffling improves the stability and reliability of the model.

REFLECTION

In a real-world platform like MeetMux, the data is not always evenly distributed. For example, if we have 10,000 users from Bangalore and only 100 users from Delhi, the dataset may end up being grouped or ordered by location.
If we don’t use shuffling in K-Fold Cross Validation, some folds might contain mostly Bangalore users while others might have very few or no Delhi users at all. This creates a biased learning situation where the model gets a lot of exposure to Bangalore user behavior but very little understanding of Delhi users.
By shuffling, each fold gets a balanced mix of users, helping the model learn general patterns and make better predictions for all users.
